```python
import json
```

```python
obj={"name":"Prithesh","age":22}
```

```python
s=json.dumps(obj,indent=4)
```

```python
py=json.loads(s)
```

```python
py
```

```python
with open("data.json","w") as f:
    json.dump(py,f)
    
```

```python
with open("data.json","r") as f:
    pd=json.load(f)
    
```
